```python
rbs['td_scorer'] = rbs['receiving_tds'] + rbs['rushing_tds'] > 0
```

```python
rbs.tail(25)
```

```python
scoring_rbs = rbs[rbs['td_scorer'] == True]
scoring_rbs = scoring_rbs[['player_name', 'team', 'week']]
scoring_rbs.head()
```

**Reasoning**:
To link scoring RBs to their home games, I will merge the `scoring_rbs` DataFrame with the `team_sched` DataFrame based on matching 'week' and where the RB's team ('team' column in `scoring_rbs`) is the 'home_team' in `team_sched`.



```python
home_scoring_rbs = pd.merge(scoring_rbs,
                            team_sched,
                            left_on=['week', 'team'],
                            right_on=['week', 'home_team'],
                            how='inner')
home_scoring_rbs.head()
```

**Reasoning**:
To aggregate the player names of scoring RBs by game, I will group the `home_scoring_rbs` DataFrame by `game_id` and then apply an aggregation to collect the `player_name` into a list for each group.



```python
scoring_rbs_by_game = home_scoring_rbs.groupby('game_id')['player_name'].apply(list).reset_index()
scoring_rbs_by_game.head()
```

```python
team_sched_with_scoring_rbs = pd.merge(team_sched,
                                      scoring_rbs_by_game,
                                      on='game_id',
                                      how='left')

team_sched_with_scoring_rbs['player_name'] = team_sched_with_scoring_rbs['player_name'].apply(lambda x: x if isinstance(x, list) else [])

team_sched_with_scoring_rbs.rename(columns={'player_name': 'scoring_rbs_home_team'}, inplace=True)
```
